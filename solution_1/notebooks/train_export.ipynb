```python
# Install required dependencies
# !pip install ultralytics onnx onnxruntime

from ultralytics import YOLO
import shutil

# Initialize the YOLOv10 Nano architecture
# YOLOv10 introduces consistent dual assignments, eliminating the NMS bottleneck natively
model = YOLO("yolov10n.pt") 

# Execute the training loop
# Adjust 'data.yaml' to point to your CVAT exported dataset paths
results = model.train(
    data="D:/ml/project/object-detection/data/dataset/data.yaml",
    epochs=100,
    imgsz=1024,
    batch=4,
    device=0, # Use GPU
    workers=2,
    project="tire_detection",
    name="run1"
)

# Export the trained model to ONNX format with FP16 quantization
# Dynamic axes allow the web implementation to accept different batch sizes or image shapes if needed
export_path = model.export(
    format="onnx",
    half=True, 
    dynamic=False, 
    simplify=True
)

print(f"Production model exported to: {export_path}")
# Move the resulting .onnx file to the web/models/ directory
```
